```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function

import copy
import sys

import math
import matplotlib.pyplot as plt
import numpy
import pandas
import sklearn.metrics
import sklearn.model_selection
import sklearn.linear_model
import sklearn.preprocessing
```

```python
data = pandas.read_csv('./HTRU_4.csv', header=None, names=['x%i' % (i) for i in range(8)] + ['y'])
data
```

```python
X = numpy.asarray(data[['x%i' % (i) for i in range(8)]])
X
```

```python
X = numpy.hstack((numpy.ones((X.shape[0],1)), X))
X
```

```python
y = numpy.asarray(data['y'])
X_train, X_test, y_train, y_test =sklearn.model_selection.train_test_split(X, y, test_size = 1 - 0.5, random_state=0)
```

```python
minmax_scaler = sklearn.preprocessing.MinMaxScaler(feature_range=(0, 1)).fit(numpy.vstack((X_train, X_test))) #将属性缩放到一个指定范围->0-1
minmax_scaler
```

```python
X_train_scale = minmax_scaler.transform(X_train)
print(X_train_scale)
X_test_scale = minmax_scaler.transform(X_test)
print(X_test_scale)
```
